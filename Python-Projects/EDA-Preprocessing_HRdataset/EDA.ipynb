#### Step 7: Finding the columns which have constant values. This is required to check becuase these are not contributing in analysis

```python
# Identifying constant columns
constant_columns = [col for col in df.columns if df[col].nunique() == 1]

# Display the constant columns
print(f"Constant columns: {constant_columns}")

# Removing constant columns
df = df.drop(columns=constant_columns)
```

###### We can see that the features 'StandardHours', 'Over18' and 'EmployeeCount' have constant values, so we can drop these values.

##### Plotting swarmplot on Overtime, Age and Marital status

```python
# Overtime Age and Marital Status

plt.figure(figsize=(10,10))

sns.swarmplot(x=df['OverTime'],y=df['Age'],hue=df['MaritalStatus'])
plt.plot()
```

#### Step 13: Group by on basis of Jobrole with avg monthly salary

```python
income_group=df.groupby(by='JobRole').agg(Mean_salary=('MonthlyIncome','mean'))
income_group
```

##### Sorting values of Mean_salary

```python
income_group.sort_values(by='Mean_salary',ascending=False)
```

#### Step 14: Visualizing the mean_salary with job role

```python
plt.figure(figsize=(5,5))

sns.barplot(x=income_group.index,y=income_group.Mean_salary, palette='Set2')
plt.xticks(rotation=90)
plt.show()
```

#### Q. With respect to deparments what is the maximum hike in salary and their max_rating
